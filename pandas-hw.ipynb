Задание 1

Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd

data_movies = pd.read_csv('movies.csv')
data_ratings = pd.read_csv('ratings.csv')
```

```python
max_rating_movies_ = data_ratings.groupby(['rating', 'movieId']).size().reset_index(name='count')
max_rating_movies_ = max_rating_movies_.sort_values(by=['rating', 'movieId', 'count'], ascending=[False, True, False])
max_rating_movies_.sort_values(by=['count'], ascending = False).head(5)
```

```python
best_rated_movie = data_movies[(data_movies['movieId'] == 318)]
best_rated_movie.head()
```

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
df_power = pd.read_csv('power.csv')
df_baltic_states = df_power[(df_power['country'] == 'Latvia') | (df_power['country'] == "Lithuania") | (df_power['country'] == 'Estonia')] 
```

```python
df_baltic_states = df_baltic_states[(df_baltic_states['category'] == 4) | (df_baltic_states['category'] == 12) | (df_baltic_states['category'] == 21)]
```

```python
df_baltic_states = df_baltic_states[(df_baltic_states['year'] == 2005) | (df_baltic_states['year'] == 2006) | (df_baltic_states['year'] == 2007) | (df_baltic_states['year'] == 2008) | (df_baltic_states['year'] == 2009) | (df_baltic_states['year'] == 2010)]
```

```python
quantity = df_baltic_states[(df_baltic_states['quantity'] >= 0)]
```

```python
quantity = quantity['quantity'].sum()
quantity
```
